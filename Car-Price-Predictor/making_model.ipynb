Now we will read our `cleanedcar.csv` and will made our model

```python
import pandas as pd
df = pd.read_csv('cleaned_car.csv')
df
```

```python
X = df.drop(labels=['Price'],axis=1)
```

```python
X
```

```python
y = df.Price
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, train_targets, val_targets = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False, drop='first'),['company', 'fuel_type', 'model'])], remainder='passthrough')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestRegressor())
])
```

```python
pipe.fit(X_train, train_targets)
```

```python
pipe.score(X_train, train_targets)
```

```python
pred = pipe.predict(X_train)
```

```python
import pickle
pickle.dump(pipe, open('model.pkl', 'wb'))
```
